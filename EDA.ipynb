```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('data/data1.csv')  
df2 = pd.read_csv('data/data2.csv')  
df3 = pd.read_csv('data/data3.csv')
```

### Histogram of Input features

```python
# Create a 1x3 subplot grid
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

display_col = "NDVI"
# Plot histograms side by side
sns.histplot(data=df1, x=display_col, bins=30, kde=True, color='skyblue', ax=axes[0],
            element='step')
axes[0].set_title(display_col)
axes[0].set_ylabel('Frequency')

sns.histplot(data=df2, x=display_col, bins=30, kde=True, color='salmon', ax=axes[1],
            element='step')
axes[1].set_title(display_col)
axes[1].set_ylabel('Frequency')

sns.histplot(data=df3, x=display_col, bins=30, kde=True, color='limegreen', ax=axes[2],
            element='step')
axes[2].set_title(display_col)
axes[2].set_ylabel('Frequency')

plt.tight_layout()  # Prevents overlapping labels
plt.show()
```
